# Data visualization 

```python
# libraries 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# loading data 
file_path = "C:/Users/Kiwi/OneDrive - Kruvelab/Yvonne/02_courses/UT machine learning/project work/data_ready_231029.csv"
df_calibrations = pd.read_csv(file_path)
df_calibrations_lab1 = df_calibrations[df_calibrations['lab']=='L1']
```

```python
# Get a list of unique compounds
compounds = df_calibrations_lab1['compound'].unique()

# Define a color palette for the 'note' variable
palette = sns.color_palette('Set1')

for compound in compounds:
    plt.figure(figsize=(8, 6))  # Set your desired figure size
    
    # Filter the data for the current compound
    compound_df = df_calibrations_lab1[df_calibrations_lab1['compound'] == compound]
    
    # Create a scatter plot
    sns.scatterplot(data=compound_df, x='c_real_M', y='peak_area', hue='note', palette=palette)
    
    # Set plot title
    plt.title(f'Scatter Plot for Compound {compound}')
    
    # Customize labels and legends as needed
    plt.xlabel('c_real_M')
    plt.ylabel('peak_area')
    plt.legend(title='Note')
    
    # Customize other plot settings as desired
    # plt.grid(True)
    # plt.tight_layout()
    
    # Save or display the plot
    # plt.savefig(f'compound_{compound}_scatter.png')  # To save the plot
    plt.show()
```
